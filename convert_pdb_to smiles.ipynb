```python
# نصب RDKit در Colab
!pip install rdkit

from rdkit import Chem
from google.colab import files
import os

# درخواست آپلود فایل PDB
print("لطفاً فایل PDB خود را آپلود کنید:")
uploaded = files.upload()

# دریافت نام فایل آپلودشده
pdb_file = list(uploaded.keys())[0]

# خواندن فایل PDB
mol = Chem.MolFromPDBFile(pdb_file, removeHs=True)

if mol is None:
    print("خطا در خواندن فایل PDB. لطفاً فایل را بررسی کنید.")
else:
    # تبدیل به SMILES
    smiles = Chem.MolToSmiles(mol, isomericSmiles=True)
    print("SMILES تولیدشده:", smiles)

    # ذخیره SMILES در فایل (اختیاری)
    with open("output_smiles.txt", "w") as f:
        f.write(smiles)
    print("SMILES در فایل 'output_smiles.txt' ذخیره شد.")
```
